```python
class Universidad:
    def __init__(self, nombre, rector, numeroFacultades, rankingMundial):
        self.nombre = nombre
        self.rector = rector
        self.numeroFacultades = numeroFacultades
        self.rankingMundial = rankingMundial

    def admitirEstudiante(self, nombre_estudiante):
        print(f"\n[{self.nombre}]: ✅ El estudiante {nombre_estudiante} ha sido admitido.")

    def investigar(self, tema):
        print(f"[{self.nombre}]: 🔬 Iniciando investigación sobre: {tema}.")

    def publicarPaper(self, titulo_paper):
        print(f"[{self.nombre}]: 📜 Paper '{titulo_paper}' publicado bajo la rectoría de {self.rector}.")

# --- Función para capturar datos por teclado ---
def crear_universidad_por_input(numero):
    print(f"\n--- Ingrese datos para la Universidad {numero} ---")
    nombre = input("Nombre de la universidad: ")
    rector = input("Nombre del rector: ")
    n_facultades = input("Número de facultades: ")
    ranking = input("Puesto en el ranking mundial: ")

    # Retornamos el objeto creado con esos datos
    return Universidad(nombre, rector, n_facultades, ranking)

# --- Creación de Objetos mediante Input ---
universidad1 = crear_universidad_por_input(1)
universidad2 = crear_universidad_por_input(2)

# --- Prueba de funciones con Inputs adicionales ---
print("\n--- Probando funciones del sistema ---")

# Ejemplo con Universidad 1
estudiante = input(f"Nombre del estudiante para {universidad1.nombre}: ")
universidad1.admitirEstudiante(estudiante)

# Ejemplo con Universidad 2
tema_inv = input(f"Tema de investigación para {universidad2.nombre}: ")
universidad2.investigar(tema_inv)

paper_titulo = input(f"Título del paper para {universidad2.nombre}: ")
universidad2.publicarPaper(paper_titulo)
print("programa finalizado")
print("programa realizado por Nicole Robles NL = 46")
```
