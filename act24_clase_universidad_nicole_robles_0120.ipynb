```python
# Definición de la clase
class Universidad:
    def __init__(self, nombre, rector, numeroFacultades, rankingMundial):
        # Atributos de la clase
        self.nombre = nombre
        self.rector = rector
        self.numeroFacultades = numeroFacultades
        self.rankingMundial = rankingMundial

    # Función 1: Admitir estudiante
    def admitirEstudiante(self, nombre_estudiante):
        print(f"[{self.nombre}]: El estudiante {nombre_estudiante} ha sido admitido satisfactoriamente.")

    # Función 2: Investigar
    def investigar(self, tema):
        print(f"[{self.nombre}]: Iniciando proyecto de investigación sobre: {tema}.")

    # Función 3: Publicar Paper
    def publicarPaper(self, titulo_paper):
        print(f"[{self.nombre}]: Se ha publicado el artículo científico titulado '{titulo_paper}' bajo la gestión del rector {self.rector}.")

# --- Creación de Objetos (Instancias) ---

# Universidad 1: Datos de ejemplo
universidad1 = Universidad(
    nombre="Universidad Nacional Autónoma",
    rector="Dr. Alejandro Magno",
    numeroFacultades=15,
    rankingMundial=120
)

# Universidad 2: Datos de ejemplo
universidad2 = Universidad(
    nombre="Instituto Tecnológico Superior",
    rector="Ing. Sofía Reyes",
    numeroFacultades=8,
    rankingMundial=450
)

# --- Ejecución de Funciones (Mensajes) ---

print("--- Actividades de la Universidad 1 ---")
universidad1.admitirEstudiante("Carlos Pérez")
universidad1.investigar("Inteligencia Artificial en Medicina")
universidad1.publicarPaper("Avances en Redes Neuronales")

print("\n--- Actividades de la Universidad 2 ---")
universidad2.admitirEstudiante("María García")
universidad2.investigar("Energías Renovables en Zonas Rurales")
universidad2.publicarPaper("Impacto de Paneles Solares Orgánicos")
```
